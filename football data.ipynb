```python
#loading required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   # used for data visualization
import seaborn as sns   # used for data visualization
```

```python
#seaborn is an extension of matplotlib
# built on matplotlib
fifa=pd.read_csv('Downloads/player_20.csv')
```

```python
fifa.head()
```

```python
# to know how many columns are present
for col in fifa.columns:
    print(col)
```

```python
# to know the records rows and columns
fifa.shape
```

```python
# how many players belong to which country
# value_counts method gives the frequency of keys and values
fifa['nationality'].value_counts()
```

```python
# to get top 10 countries
fifa['nationality'].value_counts()[0:10]
```

```python
# to get top five countries
fifa['nationality'].value_counts()[0:5]
```

```python
# to make a bar graph
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='g')
plt.show()
```

```python
 # to get the keys of top 5 countries
fifa['nationality'].value_counts()[0:5].keys()
```
